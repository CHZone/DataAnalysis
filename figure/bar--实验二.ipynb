```python
%matplotlib inline
import matplotlib.pyplot as plt  
```

```python
name_list = ['VDSM-LSTM','指数平滑','VDSM','LSTM']
# 所有线路
num_list = [4.868, 7.949, 7.303, 6.877]  
# 线路3-8
num_list1 = [4.078,7.505, 7.401, 6.77]  
# 线路1-2、9-10
num_list2 = [5.552, 8.615, 7.106, 7.037]
X =list(range(len(num_list)))  

plt.figure(figsize=(9,5)) 
# 绘制所有线路
plt.bar(X, num_list, width=0.22, label='线路1-10',tick_label = name_list,facecolor = 'lightskyblue',edgecolor='white')  
#  添加数字
for x,y in zip(X,num_list):
    plt.text(x+0.12,y+0.05,'%.3f' %y, ha='center')
    
for i in range(len(X)):  
    X[i] = X[i] + 0.22  

# 绘制线路3-8
plt.bar(X, num_list1, width=0.22, label='线路3-8',tick_label = name_list,facecolor = 'yellowgreen',edgecolor='white')  

for x,y in zip(X,num_list1):
    plt.text(x+0.12,y+0.05,'%.3f' %y, ha='center')
    
for i in range(len(X)):  
    X[i] = X[i] + 0.22 
    
# 绘制线路1-2， 9-10
plt.bar(X, num_list2, width=0.22, label='线路1-2,9-10',tick_label = name_list,facecolor = 'yellow',edgecolor='white')  

for x,y in zip(X,num_list2):
    plt.text(x+0.12,y+0.05,'%.3f' %y, ha='center')
plt.legend(loc='lower right')  

plt.ylabel('MAPE(%)')
plt.xlabel('模型')
plt.title('各模型预测结果的 MAPE(%)')
plt.show()
```
